```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# Step 1: Load and explore the data
def load_data(filepath):
    """Load precious metals dataset"""
    df = pd.read_csv(filepath)
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date')
    return df
```

```python
# Step 2: Prepare data for LSTM
def create_sequences(data, seq_length):
    """
    Create sequences for LSTM training
    seq_length: number of time steps to look back
    """
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)
```

```python
# Step 3: Preprocess the data
def preprocess_data(df, target_metal='Gold', seq_length=60, train_split=0.8):
    """
    Preprocess data for LSTM training
    target_metal: 'Gold', 'Silver', 'Platinum', or 'Palladium'
    seq_length: number of previous time steps to use for prediction
    """
    # Filter for the target metal and select the 'close' price
    metal_df = df[df['commodity'] == target_metal].copy()
    if metal_df.empty:
        raise ValueError(f"No data found for commodity: {target_metal}")

    data = metal_df['close'].values.reshape(-1, 1)

    # Normalize the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)

    # Create sequences
    X, y = create_sequences(scaled_data, seq_length)

    # Split into train and test sets
    split_idx = int(len(X) * train_split)
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    return X_train, X_test, y_train, y_test, scaler
```

```python
# Step 4: Build LSTM model
def build_lstm_model(seq_length, n_features=1):
    """
    Build LSTM model architecture
    seq_length: number of time steps in input
    n_features: number of features (1 for univariate)
    """
    model = Sequential([
        # First LSTM layer with return sequences
        LSTM(units=50, return_sequences=True,
             input_shape=(seq_length, n_features)),
        Dropout(0.2),

        # Second LSTM layer
        LSTM(units=50, return_sequences=True),
        Dropout(0.2),

        # Third LSTM layer
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),

        # Dense layers
        Dense(units=25),
        Dense(units=1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error',
                  metrics=['mae'])
    return model

```

```python
# Step 5: Train the model
def train_model(model, X_train, y_train, X_test, y_test, epochs=50, batch_size=32):
    """Train LSTM model with callbacks"""

    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )

    model_checkpoint = ModelCheckpoint(
        'best_lstm_model.keras',
        monitor='val_loss',
        save_best_only=True
    )

    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        callbacks=[early_stopping, model_checkpoint],
        verbose=1
    )

    return history

```

```python
# Step 6: Make predictions and evaluate
def evaluate_model(model, X_test, y_test, scaler):
    """Evaluate model and calculate metrics"""

    # Make predictions
    predictions = model.predict(X_test)

    # Inverse transform to get actual prices
    predictions = scaler.inverse_transform(predictions)
    y_test_actual = scaler.inverse_transform(y_test)

    # Calculate metrics
    mse = mean_squared_error(y_test_actual, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_actual, predictions)
    mape = np.mean(np.abs((y_test_actual - predictions) / y_test_actual)) * 100

    print(f"Model Performance:")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    print(f"MAPE: {mape:.2f}%")

    return predictions, y_test_actual

```

```python
def plot_results(history, predictions, actual, title):
    # Model loss plot REMOVED

    # Price prediction plot (KEEP this)
    plt.figure(figsize=(12, 4))
    plt.plot(actual, label='Actual Price')
    plt.plot(predictions, label='Predicted Price')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Main execution workflow
def main():
    # Configuration
    FILEPATH = 'all_commodities_data.csv'  # Update with your file path
    TARGET_METAL = 'Silver'  # Choose: 'Gold', 'Silver', 'Platinum', 'Palladium'
    SEQ_LENGTH = 60  # Use 60 previous time steps
    EPOCHS = 50
    BATCH_SIZE = 32

    # Load data
    print("Loading data...")
    df = load_data(FILEPATH)
    print(f"Data shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")

    # Preprocess data
    print("\nPreprocessing data...")
    X_train, X_test, y_train, y_test, scaler = preprocess_data(
        df, TARGET_METAL, SEQ_LENGTH
    )
    print(f"Training samples: {X_train.shape[0]}")
    print(f"Testing samples: {X_test.shape[0]}")

    # Build model
    print("\nBuilding LSTM model...")
    model = build_lstm_model(SEQ_LENGTH)
    model.summary()

    # Train model
    print("\nTraining model...")
    history = train_model(model, X_train, y_train, X_test, y_test,
                         epochs=EPOCHS, batch_size=BATCH_SIZE)

    # Evaluate model
    print("\nEvaluating model...")
    predictions, actual = evaluate_model(model, X_test, y_test, scaler)

    # Visualize results
    plot_results(history, predictions, actual,
                title=f'{TARGET_METAL} Price Prediction')

    # Save the model
    model.save('lstm_precious_metals_final.keras')
    print("\nModel saved successfully!")

if __name__ == "__main__":
    main()
```
